Ejercicio:Distribucion Poisson
Si en central telefonica recibe en promedio 4 llamadaspor hora, calcular las siguientes probabilidades :
A-¿Que en una hora se reciba una llamada
B-¿Que en una hora se reciban 3 llamadas
C-¿Que en una hora se reciba, al menos,una llamada?
D-¿Que en una hora se reciban, como mucho, 4 llamadas?
E-Si la frecuencia que recibe las llamadas es relativamente constante es deci,se mantiene constante el promedio de llamadas recibidas por hora, clacular la probababilidad que en dos horas se reciban exactamente 9 llamadas

Resolucion:
Sea:
λ = 4 llamadas por hora
x = 1 llamada
e = la base del logaritmo natural
x! = el factorial de x


```python
import math
import matplotlib.pyplot as plt
```

resultado_pow = math.pow(base, exponente) Potencia
resultado_factorial = math.factorial(numero) Factorial

```python
def poisson(λ,x):
    #obtener euler 
    e=math.e
    λ_resultado=math.pow(λ,x)
    e_resultado=math.pow(e,(-λ))
    x_factorial=math.factorial(x)
    fx=(λ_resultado*e_resultado)/(x_factorial)
    return fx
```

```python
#poisson(Promedio,tasa ocurra)
fx1=poisson(4,1)
fx2=poisson(4,3)
```

```python
#A-¿Que en una hora se reciba una llamada
#B-¿Que en una hora se reciban 3 llamadas
print("De que se reciba una llamada:",fx1)
print("De que se reciba 3  llamadas:",fx2)
```

```python
#al menos una llamada
#C-¿Que en una hora se reciba, al menos,una llamada?
fx3=poisson(4,0)
print(1-fx3)
 
```

```python
def probabilidadcombinada(rango):
    #D-¿Que en una hora se reciban, como mucho, 4 llamadas?
    resultados=[]
    sumatoria=0
    for i in range(rango):
        resultados.append(poisson(4,i))
        sumatoria+=resultados[i]
    return resultados,sumatoria
```

```python
resultados,sumatoria=probabilidadcombinada(12)
```

```python
plt.bar(range(12), resultados, color='blue', alpha=0.7)

# Añadir etiquetas y título
plt.xlabel('Número de llamadas en una hora')
plt.ylabel('Probabilidad')
plt.title('Poisson distribution Mean=4')

# Mostrar el histograma
plt.show()
```
